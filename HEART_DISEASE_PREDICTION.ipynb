```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv('HeartDiseaseTrain-Test.csv')
heart_data.drop
```

```python
heart_data.head()
```

```python

label_encoder = LabelEncoder()
heart_data['chest_pain_type'] = label_encoder.fit_transform(heart_data['chest_pain_type'])
heart_data['rest_ecg'] = label_encoder.fit_transform(heart_data['rest_ecg'])
heart_data['exercise_induced_angina'] = label_encoder.fit_transform(heart_data['exercise_induced_angina'])
heart_data['slope'] = label_encoder.fit_transform(heart_data['slope'])
heart_data['thalassemia'] = label_encoder.fit_transform(heart_data['thalassemia'])
heart_data['fasting_blood_sugar'] = label_encoder.fit_transform(heart_data['fasting_blood_sugar'])
heart_data['vessels_colored_by_flourosopy'] = label_encoder.fit_transform(heart_data['vessels_colored_by_flourosopy'])
heart_data['sex'] = label_encoder.fit_transform(heart_data['sex'])
```

```python
heart_data
```

```python
heart_data.isnull().sum()
```

```python
x = heart_data.drop(columns='target', axis=1)
y = heart_data['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
print(x.shape,x_train.shape,x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
